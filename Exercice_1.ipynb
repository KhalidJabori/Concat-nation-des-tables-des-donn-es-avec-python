# <span style="color:red"> <span  style="font-family:Times New Roman"> *Importing libraries* </span> </span></b> 


```python
import pandas as pd
import numpy as np
```

# <b><span style="color:red"> <span  style="font-family:Times New Roman"> *_Importing Data_* </span> </span></b>

## <span style="color:blue"> *Table 1*</span>


```python
tache_11=pd.read_csv("NE_A_Br.csv")
print(tache_11.dtypes)
print(tache_11.shape)
tache_1=tache_11.drop(columns=["Unnamed: 4",  "Unnamed: 5", "Unnamed: 6"])
tache_1
```

## <span style="color:blue"> Table 2  </span>


```python
tache_22=pd.read_csv('T1_06_07_2022__10_31_51.csv')
tache_2=tache_22.drop(columns=["Unnamed: 10",  "Unnamed: 11", "Unnamed: 12"])
tache_2=tache_2.drop(index=[0,1,2])
tache_2.columns = list(tache_2.loc[3])
tache_2=tache_2.drop(index=[3])
tache_2
```

## <span style="color:orange"> Operations on table 2  </span>


```python
tache_2[["Année","Mois","NE","CP","QM Courante","Q Correspondante","Var1 en %","Q Précédente","Var2 en %"]] = tache_2[["Année","Mois","NE","CP","QM Courante","Q Correspondante","Var1 en %","Q Précédente","Var2 en %"]].apply(pd.to_numeric)
print(tache_2.dtypes)
print(tache_2.shape)
tache_2
```

# <span style="color:red"> <span  style="font-family:Times New Roman"> *Joining Tables* </span> </span></b> 

```python
tab_1=tache_2.merge(tache_1,on='NE')
tab_1.sort_values("NE")
tab_1.reset_index()
tab_1.index_col=0
tab_1
```

# <span style="color:red"> <span  style="font-family:Times New Roman"> *Exploring Data* </span> </span></b> 

```python
tab_1_F=tab_1.to_excel('Exemple_F_T1_06_07_2022__10_31_51.xlsx',encoding="utf-8",engine="xlsxwriter",index=False)
tab_1_F
```
